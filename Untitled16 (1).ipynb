```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

imagem = cv2.imread("/content/one-piece-anime-1015-luffy-chapeu-postcover.jpg")

(alt, lar) = imagem.shape[:2]#captura altura e largura
centro = (lar // 2, alt // 2 )#acha centro

M = cv2.getRotationMatrix2D(centro, 30, 1.0) #30 graus

img_rotacionada = cv2.warpAffine(imagem, M, (lar, alt))

cv2_imshow(img_rotacionada)
```
