```python
from google.colab.patches import cv2_imshow
import cv2

imagem = cv2.imread("/content/one-piece-anime-1015-luffy-chapeu-postcover.jpg")

#uma maneira
#Lnova = 100
#Anova = 100

largura = imagem.shape[1]
altura = imagem.shape[0]
propocao = float(altura/largura)
Lnova = 150
Anova = int(Lnova*propocao)

imagem = cv2.resize(imagem, (Lnova, Anova))


cv2_imshow(imagem)
```
