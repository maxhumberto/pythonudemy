```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

imagem = cv2.imread("/content/one-piece-anime-1015-luffy-chapeu-postcover.jpg")

cv2_imshow(imagem)
largura = imagem.shape[1]
altura = imagem.shape[0]
propocao = float(altura/largura)
Lnova = 320
Anova = int(Lnova*propocao)

Tnovo=(Lnova, Anova)

# interpolation serve para preencher os pixes que na nova imagem
img_redimensionada = cv2.resize(imagem, Tnovo, interpolation = cv2.INTER_AREA)

imagem = cv2.resize(imagem, (Lnova, Anova))

cv2_imshow(img_redimensionada)
cv2_imshow(imagem)
```
